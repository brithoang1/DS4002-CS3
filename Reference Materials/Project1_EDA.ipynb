## Section 1 - Data Cleaning

## Section 2 - EDA

```python
# linegraph of tweet volume over time

# convert 'created_at' times to datetime format
trump_df['created_at'] = pd.to_datetime(trump_df['created_at'])
biden_df['created_at'] = pd.to_datetime(biden_df['created_at'])

# resample data to calculate total number of tweets by the hour
tweet_volume_trump = trump_df.resample('H', on='created_at').size()
tweet_volume_biden = biden_df.resample('H', on='created_at').size()

# combine tweet volumes into single df and fill na values with 0 to have complete records
tweet_volume = pd.DataFrame({
    'Trump': tweet_volume_trump,
    'Biden': tweet_volume_biden
}).fillna(0)

# plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=tweet_volume, x=tweet_volume.index, y='Trump', color='red', label='Trump', linestyle='-')
sns.lineplot(data=tweet_volume, x=tweet_volume.index, y='Biden', color='blue', label='Biden', linestyle='-')
plt.title('Tweet Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Tweets')
plt.legend(title='Candidate')
plt.xticks(rotation=45)
plt.show
```

```python
# bargraph of tweets by state

# define list of state abbreviations
us_states = [
    'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI',
    'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI',
    'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC',
    'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT',
    'VT', 'VA', 'WA', 'WV', 'WI', 'WY'
]

# group tweets by state and count the number of tweets in each
trump_counts = trump_df[trump_df['state_code'].isin(us_states)].groupby('state_code').size().reset_index(name='tweet_count')
biden_counts = biden_df[biden_df['state_code'].isin(us_states)].groupby('state_code').size().reset_index(name='tweet_count')

# combine tweets counts using 'state_code'
combined_counts = pd.merge(trump_counts, biden_counts, on='state_code', how='outer', suffixes=('_Trump', '_Biden')).fillna(0)
melted_counts = combined_counts.melt(id_vars='state_code', value_vars=['tweet_count_Trump', 'tweet_count_Biden'], var_name='Candidate', value_name='tweet_count')

# rename candidates to 'Trump' and 'Biden'
melted_counts['Candidate'] = melted_counts['Candidate'].replace({'tweet_count_Trump': 'Trump', 'tweet_count_Biden': 'Biden'})

# plot
plt.figure(figsize=(14, 8))
palette = {'Trump': 'red', 'Biden': 'blue'}
sns.barplot(data=melted_counts, x='state_code', y='tweet_count', hue='Candidate', palette=palette)
plt.title('Tweet Volume by State')
plt.xlabel('State Code')
plt.ylabel('Number of Tweets')
plt.legend(title='Candidates')
plt.xticks(rotation=45)
plt.show()
```

```python
# word clouds of most common words for each candidate

import re
from wordcloud import WordCloud

# function to create plot for both candidates
def generate_wordcloud(tweets, candidate_name):
    # combine all tweets into single string and drop na values
    all_tweets = ' '.join(tweets.dropna())

    # define words to exclude
    excluded_words = [candidate_name, 'DonaldTrump', 'JoeBiden', 'realDonaldTrump']

    # remove URLs and excluded words
    pattern = r'http\S+|www\S+|https\S+|{}'.format('|'.join(excluded_words))
    all_tweets = re.sub(pattern, '', all_tweets, flags=re.MULTILINE | re.IGNORECASE)

    # generate the word cloud
    wordcloud = WordCloud(width=800, height=400,
                          background_color='white',
                          colormap='viridis').generate(all_tweets)

    # plot
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud of {candidate_name}\'s Tweets', fontsize=16)
    plt.show()

generate_wordcloud(trump_df['tweet'], 'Trump')
generate_wordcloud(biden_df['tweet'], 'Biden')
```
